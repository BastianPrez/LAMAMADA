# 1. Indice de masa corporal
[Contenido](#Contenido)

El 칤ndice de masa corporal es un intento de cuantificar la cantidad de masa tisular (m칰sculo, grasa y hueso) de un individuo y luego clasificar a esa persona como con bajo peso, peso normal, sobrepeso u obesidad seg칰n ese valor.

El 칤ndice de masa corporal de una persona con peso w (en kilogramos) y altura h (en metros) se calcula como:

$$\frac{w}{h^2}$$

Utilice la f칩rmula anterior para calcular el 칤ndice de masa corporal de alguien que pesa 66.2 kilogramos y mide 1.82 metros. Imprime el resultado en la pantalla.

````python
# Escribe tu codigo debajo

peso = 66.2
altura = 1.82

imc = peso / (altura ** 2)

print ("El 칤ndice de masa corporal es:", imc)
````

# 5. Celsius y Fahrenheit
[Contenido](#Contenido)

Diferentes regiones del mundo utilizan diferentes unidades para medir la temperatura. En este problema de pr치ctica, convertiremos temperaturas expresadas en Celsius (춿C) a Fahrenheit (춿F).

Para convertir una temperatura de Celsius a Fahrenheit, empleamos

$$^{\circ} F = ^{\circ} C \cdot 1.8 + 32$$

Por ejemplo, `10춿C` es igual a `50춿F`:

$$^{\circ} F = 10 \cdot 1.8 + 32 = 18 + 32 = 50$$

1. Imprime en la pantalla el resultado de convertir `30` grados Celsius a Fahrenheit.
2. Imprime en la pantalla el resultado de convertir `86` grados Fahrenheit a Celsius.

````python
# Escribe tu codigo debajo
# Conversi칩n de Celsius a Fahrenheit
celsius = 30
fahrenheit = celsius * 1.8 + 32
print("30 grados Celsius son", fahrenheit, "grados Fahrenheit.")

# Conversi칩n de Fahrenheit a Celsius
fahrenheit = 86
celsius = (fahrenheit - 32) / 1.8
print("86 grados Fahrenheit son", celsius, "grados Celsius.")
````
